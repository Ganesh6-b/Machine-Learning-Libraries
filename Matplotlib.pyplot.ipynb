```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.array([1,2,3,4,5])
y = np.random.randn(5)
```

```python
#functional plots
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("first plot")
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
#for subplots
plt.subplot(1,3,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method
fig = plt.figure(figsize = (4,4), facecolor = 'pink')
axes = fig.add_axes([1,1,1,1])
axes.set_xlabel("x")
axes.set_ylabel('y')
axes.set_title("oom plot")
axes.plot(x,y,'r')
```

```python
fig = plt.figure(facecolor ='b')
axes1 = fig.add_axes([0.1,0.1,1.1,1.1])
axes2 = fig.add_axes([0.2,0.2,0.5,.4])
axes1.plot(x,x**2,'r', label = "red")
axes1.legend()

axes2.plot(x,y**3,'b', label = "blue")
plt.legend()
```

```python
fig,axes = plt.subplots(nrows = 2, ncols =1, facecolor = "pink")
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'b')
plt.tight_layout
plt.savefig("firfigure.png", dpi = 50)
```
